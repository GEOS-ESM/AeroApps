# Clould Effective Radius vs Aerosol Concentrations

Aerosol concentrations have been calculataded as 

* Average with PBL
* Average above PBL
* At layer nearest to cloud top

The following filter is being applied

* Eliminate multi-layer clouds
* Keep only r$_E$ retrievals with uncertainty less than 10%.


Let's consider data for a single day with optical property data thinned at 5km,

```python
from glob import glob
from renn import RENN
Files = sorted(glob('/nobackup/MODIS/051/Level2/MOD06/2012/228/MOD06_L2.A2012228.*.hdf'))
m = RENN(Files,Verb=1)             
```

Next load the pre-computed aerosol concentrations,

```python
m.loadAero('aeroFeatures.npz')
```

The simplest to depict how increased aersol concentration limits the cloud effective radius is by means of a basic scatter diagram

### Ice Clouds

```python
figure(figsize=(12,9))

subplot(221)
plot(m.OCLEV[m.iIce],m.RE[m.iIce],'go')
xlabel(r'Aerosol Concetration [g/cm$^3$]')
ylabel(r'r$_E$ [$\mu$m]')
title('OC at Ice Cloud Top')
grid()

subplot(222)
plot(m.SSLEV[m.iIce],m.RE[m.iIce],'bo')
xlabel(r'Aerosol Concetration [g/cm$^3$]')
ylabel(r'r$_E$ [$\mu$m]')
title('Sea Salt at Ice Cloud Top')
grid()

subplot(223)
plot(m.DULEV[m.iIce],m.RE[m.iIce],'ro')
xlabel(r'Aerosol Concetration [g/cm$^3$]')
ylabel(r'r$_E$ [$\mu$m]')
title('Dust at Ice Cloud Top')
grid()

subplot(224)
plot(m.SULEV[m.iIce],m.RE[m.iIce],'yo')
xlabel(r'Aerosol Concetration [g/cm$^3$]')
ylabel(r'r$_E$ [$\mu$m]')
title('Sulfates at Ice Cloud Top')
grid()
```

This does not seem to work too well, probably because of the way MODIS Collection handles cloud top pressure for low clouds. Let's see if we find any relationship with aerosol concentrations within the PBL,
